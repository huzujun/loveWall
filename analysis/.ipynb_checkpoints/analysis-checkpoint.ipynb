```python
import pkuseg
from collections import Counter
import pprint

content = ""
with open("list.txt", encoding="utf-8") as f:
    content = f.read()

seg = pkuseg.pkuseg(user_dict='my_dict.txt')
text = seg.cut(content)


stopwords = []

with open("stopword.txt", encoding="utf-8") as f:
    stopwords = f.read()

new_text = []

for w in text:
    if w not in stopwords:
        new_text.append(w)

counter = Counter(new_text)
pprint.pprint(counter.most_common(20))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
cut_text = " ".join(new_text)
cloud = WordCloud(
        #设置字体，不指定就会出现乱码
        font_path="SIMYOU.TTF",
        #font_path=path.join(d,'simsun.ttc'),
        #设置背景色
        background_color='white',
        max_words=200,
        max_font_size= 280,
        random_state=50,
        width = 1500,
        height = 1000
    )
word_cloud = cloud.generate(cut_text) # 产生词云
#word_cloud.to_file("cloudword1.jpg") #保存图片
#  显示词云图片
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```

```python
boy_words = ["小哥哥", "男孩", "男神", "男生"] 
girl_words = ["小姐姐", "女孩", "女神", "女生"]
lines = []
with open("list.txt", encoding="utf-8") as f:
    lines = f.readlines()
```

```python
boy_text = []
for line in lines:
    text = seg.cut(line)

    flag = False
    for w in text:
        if w in boy_words:
            flag = True
            break
    if flag:
        for w in text:
            boy_text.append(w)
            
girl_text = []
for line in lines:
    text = seg.cut(line)

    flag = False
    for w in text:
        if w in girl_words:
            flag = True
            break
    if flag:
        for w in text:
            girl_text.append(w)
```

```python
import numpy as np
from os import path
import os
from PIL import Image

d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
heart_mask = np.array(Image.open(path.join(d, "heart.jpg")))
meanless_word = ["表白","喜欢","希望"]
```

```python
new_text = []
for w in boy_text:
    if w not in stopwords and w not in boy_words and w not in girl_words and w not in meanless_word:
        new_text.append(w)   
cut_text = " ".join(new_text)
cloud = WordCloud(
        #设置字体，不指定就会出现乱码
        font_path="SIMYOU.TTF",
        #font_path=path.join(d,'simsun.ttc'),
        #设置背景色
        background_color='white',
        max_words=150,
        max_font_size= 150,
        mask = heart_mask,
        random_state=50
    )
word_cloud = cloud.generate(cut_text) # 产生词云
word_cloud.to_file("boy_words.jpg") #保存图片
#  显示词云图片

plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()

            
counter = Counter(new_text)
pprint.pprint(counter.most_common(50))
```

```python
new_text = []
for w in girl_text:
    if w not in stopwords and w not in boy_words and w not in girl_words and w not in meanless_word:
        new_text.append(w)    
cut_text = " ".join(new_text)
cloud = WordCloud(
        #设置字体，不指定就会出现乱码
        font_path="SIMYOU.TTF",
        #font_path=path.join(d,'simsun.ttc'),
        #设置背景色
        background_color='white',
        max_words=150,
        max_font_size= 150,
        mask = heart_mask,
        random_state=50,
    )
word_cloud = cloud.generate(cut_text) # 产生词云
word_cloud.to_file("girl_words.jpg") #保存图片
#  显示词云图片
plt.imshow(word_cloud)
plt.axis('off')
plt.show()

            
counter = Counter(new_text)
pprint.pprint(counter.most_common(50))
```

```python
majors = ["电信","机电","经管","交运","运输","法院","法学院","理院","理学院","计院","计算机","电气","软件","软院","建艺","语传"]
new_text = []

for w in girl_text:
    if w in majors:
        new_text.append(w)

counter = Counter(new_text)
majors_2 = {}
for c in counter:
    majors_2[c] = 0
for c in counter:
    if (c=='软院'): majors_2['软件']+=counter[c]
    elif (c=='计院'): majors_2['计算机']+=counter[c]
    elif (c=='交运'): majors_2['运输']+=counter[c]
    elif (c=='法院'): majors_2['法学院']+=counter[c]
    else: majors_2[c]+=counter[c]
name = []
count = []
for c in counter:
    if majors_2[c]!=0:
        print(c, majors_2[c])
        name.append(c)
        count.append(majors_2[c])
x = np.array(name)
y = np.array(count)
import seaborn as sns
plt.rcParams['font.sans-serif']=['simhei'] #用来正常显示中文标签 
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
sns.barplot(x, y)
plt.xlabel('专业')
plt.ylabel('次数')
#plt.ylim(49, 54)
plt.title("各专业女生曝光数")
plt.savefig("girl_count_majors.jpg")
plt.show()
```
